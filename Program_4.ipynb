```python
arr = []
```

```python
def poly():
    while True:
        power = int(input("Enter the power : "))
        coeff = int(input("Enter the coeffecient : "))
        term = []
        term.append(power)
        term.append(coeff)
        arr.append(term)
        if power == 0:
            break
```

```python
poly()
```

```python
def func(x):
    fun = 0
    for term in arr:
        fun = fun + (term[1]*pow(x,term[0]))
        
    return fun
```

```python
def hill_climbing_search(x0, delta, tol=1e-6, max_iter=10000):
    x = x0
    i = 0

    while True:
        i += 1

        x_up = x + delta
        x_down = x - delta

        f_x_up = func(x_up)
        f_x_down = func(x_down)

        if i > max_iter:
            print("Max iterations reached.")
            break

        if f_x_up < f_x_down:
            x = x_down
        else:
            x = x_up

        if abs(x - x0) < tol:
            break

        x0 = x

    return x
```

```python
import random
x0 = float(input("Enter the values of x0 : "))
maxis = []
deltas = []
for i in range(0,10):
    delta = random.random()
    deltas.append(delta)
    x_max = hill_climbing_search(x0,delta)
    maxis.append(x_max)
    print(f"The value of x that maximises the function is {x_max}.")
```

```python
import matplotlib.pyplot as plt

plt.plot(maxis)
```
